```python
moviedf['release date'] = pd.to_datetime(moviedf['release date'],errors='coerce')
#moviedf.head()
# releasedf = moviedf.set_index('release date').drop(columns=['movie id', 'movie title'])
# releasedf.index = releasedf.index.year
# newrel = releasedf.groupby(releasedf.index)
# newrel.head(10)

releasedf = moviedf.drop(columns=['movie id', 'movie title'])
#releasedf.head()
releasedf['release date']=releasedf['release date'].dt.year
newrel = releasedf.groupby(['release date'])
newrelsum = newrel.sum()
#newrelsum.head()
newrelsum['total'] = newrelsum.sum(axis=1)
#newrelsum.head()
newrelperc = newrelsum.div(newrelsum['total'], axis= 0) * 100
newrelperc.head()
areaplot = newrelperc.plot(figsize=(20, 20), subplots=True, ylim=(0,100))
# # fig,ax = plt.subplots(figsize=(10, 10)) 
# # sns.heatmap(newrelsum.corr(), ax=ax,annot=True, linewidths=.05, fmt= '.2f', center = 1 )
plt.show()


```

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here
grpdf1 = ratingdf.groupby(by=['movie id'])['rating'].count().rename('rating_count')
grpdf2 = ratingdf.groupby(by=['movie id'])['rating'].mean().rename('rating_mean')
newmovdf = moviedf[['movie id','movie title']].copy()

newset1 = pd.merge(grpdf1, grpdf2, how = 'outer', on ='movie id')
newset = pd.merge(newmovdf, newset1, how='outer', on ='movie id')

sliceddf = newset[newset.rating_count > 100]
arrangeddf = sliceddf.sort_values(by=['rating_mean'],ascending=False)
arrangeddf.head(25)
```
